### SAS program = Data Step + Proc Step

#### Cars Dataset

```sas
/* Data step */
data europeancars;
	set sashelp.cars;
	where origin="Europe";
run;


/* Proc step */
proc print data=europeancars;
run;
```

#### Heart Dataset

```sas
data highchol;
	set sashelp.heart;
	where Chol_Status="High";
run;

proc print data=highchol;
run;
```
