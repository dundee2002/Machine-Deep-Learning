```sas
ods noproctitle;
ods graphics / imagemap=on;

proc means data=SASHELP.FISH chartype mean std min max n vardef=df;
	var Weight;
run;
```

```sas
ods graphics / reset width=6.4in height=4.8in imagemap;

proc sgplot data=SASHELP.FISH;
	hbar Species /;
	xaxis grid;
run;

ods graphics / reset;
```
