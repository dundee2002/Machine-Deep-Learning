```sas
proc sql noprint;
    create table WORK.filter_basic as select * from sashelp.baseball where (Salary LT 750);
quit;

proc print data=WORK.filter_basic(obs=10);
    title "Filtered data set - WORK.filter_basic";
run;

title;
```

```sas
proc sql noprint;
	create table WORK.filter_basic as select  League, Name, Salary, Team from SASHELP.BASEBALL where(Salary 
		LT 750);
quit;

proc print data=WORK.filter_basic(obs=10);
	title "Filtered data set - WORK.filter_basic";
run;

title;
```

```sas
proc sql noprint;
	create table WORK.filter_compound as select * from SASHELP.BASEBALL 
		where(Salary LT 750 AND League EQ "American");
quit;

proc print data=WORK.filter_compound;
	title "Filtered data set - WORK.filter_compound";
run;

title;
```
